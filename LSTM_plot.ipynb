```python
training_loss = []
validation_loss = []
f = open('lstm_results_2.txt')
tr = []
while True:
    line = f.readline().rstrip('\n')
    if line == 'ready':
        continue
    if line.startswith('count'):
        tr.append(float(line[-7:]))
    if line.startswith('validation'):
        validation_loss.append(float(line[17:]))
    if line.startswith('epoch'):
        training_loss.append(tr)
        tr = []
    if not line:
        break
    
f.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = []
for t in training_loss:
    train += t

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(np.arange(len(train)) / len(training_loss[0]),train,label='training loss')
ax.plot(np.arange(1,len(validation_loss)+1) ,validation_loss,label='validation loss')

ax.set_xlabel('epoch')
ax.set_ylabel('loss value')
ax.legend()
```
